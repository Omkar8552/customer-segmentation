```python
# Customer Segmentation Analysis - Part 1: Data Loading & Cleaning

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# ============================================
# STEP 1: LOAD THE DATA
# ============================================
print("Loading dataset...")
df = pd.read_excel('data/Online Retail.xlsx')

print(f"Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print("\n" + "="*50)

# ============================================
# STEP 2: INITIAL EXPLORATION
# ============================================
print("\nFirst 5 rows:")
print(df.head())

print("\n" + "="*50)
print("\nDataset Info:")
print(df.info())

print("\n" + "="*50)
print("\nBasic Statistics:")
print(df.describe())

print("\n" + "="*50)
print("\nMissing Values:")
print(df.isnull().sum())

print("\n" + "="*50)
print("\nUnique Values:")
print(f"Unique Customers: {df['CustomerID'].nunique()}")
print(f"Unique Products: {df['StockCode'].nunique()}")
print(f"Unique Invoices: {df['InvoiceNo'].nunique()}")
print(f"Countries: {df['Country'].nunique()}")

# ============================================
# STEP 3: DATA CLEANING
# ============================================
print("\n" + "="*50)
print("\nSTARTING DATA CLEANING...")
print("="*50)

# 3.1: Remove rows with missing CustomerID
print(f"\nBefore removing NaN CustomerIDs: {len(df)} rows")
df = df[df['CustomerID'].notna()]
print(f"After removing NaN CustomerIDs: {len(df)} rows")

# 3.2: Remove cancelled orders (InvoiceNo starting with 'C')
print(f"\nBefore removing cancelled orders: {len(df)} rows")
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
print(f"After removing cancelled orders: {len(df)} rows")

# 3.3: Remove negative quantities and prices
print(f"\nBefore removing negative values: {len(df)} rows")
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
print(f"After removing negative values: {len(df)} rows")

# 3.4: Create TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
print("\nCreated 'TotalPrice' column")

# 3.5: Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print("Converted 'InvoiceDate' to datetime format")

# 3.6: Check date range
print(f"\nDate Range: {df['InvoiceDate'].min()} to {df['InvoiceDate'].max()}")

# ============================================
# STEP 4: EXPLORATORY DATA ANALYSIS
# ============================================
print("\n" + "="*50)
print("EXPLORATORY DATA ANALYSIS")
print("="*50)

# 4.1: Top 10 Countries by Number of Customers
print("\nTop 10 Countries by Customer Count:")
country_customers = df.groupby('Country')['CustomerID'].nunique().sort_values(ascending=False).head(10)
print(country_customers)

# 4.2: Revenue by Country (Top 10)
print("\nTop 10 Countries by Revenue:")
country_revenue = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)
print(country_revenue)

# 4.3: Most Purchased Products (Top 10)
print("\nTop 10 Most Purchased Products:")
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
print(top_products)

# ============================================
# STEP 5: SAVE CLEANED DATA
# ============================================
print("\n" + "="*50)
print("Saving cleaned data...")
df.to_csv('data/cleaned_online_retail.csv', index=False)
print("Cleaned data saved to 'data/cleaned_online_retail.csv'")

print("\n" + "="*50)
print("DATA CLEANING COMPLETE!")
print(f"Final dataset shape: {df.shape}")
print("="*50)

# Display final sample
print("\nCleaned Data Sample:")
print(df.head(10))
```

```python
# ============================================
# PHASE 3: RFM ANALYSIS
# ============================================

print("="*60)
print("STARTING RFM ANALYSIS")
print("="*60)

# Set the reference date (1 day after the last transaction)
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
print(f"\nSnapshot Date (Reference): {snapshot_date}")

# Calculate RFM metrics for each customer
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',  # Frequency
    'TotalPrice': 'sum'  # Monetary
})

# Rename columns
rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalPrice': 'Monetary'
}, inplace=True)

print(f"\nRFM Table Created!")
print(f"Number of customers: {len(rfm)}")
print("\n" + "="*60)

# Display RFM statistics
print("\nRFM Statistics:")
print(rfm.describe())

print("\n" + "="*60)
print("\nSample RFM Data (First 10 customers):")
print(rfm.head(10))

# ============================================
# RFM SCORING (1-5 scale)
# ============================================
print("\n" + "="*60)
print("CALCULATING RFM SCORES...")
print("="*60)

# Create RFM scores using quantiles (1=worst, 5=best)
# For Recency: Lower is better (recent purchase)
# For Frequency & Monetary: Higher is better

rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5])

# Convert scores to integers
rfm['R_Score'] = rfm['R_Score'].astype(int)
rfm['F_Score'] = rfm['F_Score'].astype(int)
rfm['M_Score'] = rfm['M_Score'].astype(int)

# Calculate total RFM score
rfm['RFM_Score'] = rfm['R_Score'] + rfm['F_Score'] + rfm['M_Score']

# Create RFM segment string (e.g., "555" = best customers)
rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

print("\nRFM Scores calculated successfully!")
print("\nSample RFM Data with Scores:")
print(rfm.head(10))

# ============================================
# CUSTOMER SEGMENTATION BASED ON RFM
# ============================================
print("\n" + "="*60)
print("CREATING CUSTOMER SEGMENTS...")
print("="*60)

# Define segment names based on RFM scores
def segment_customers(df):
    if df['RFM_Segment'] == '555':
        return 'Champions'
    elif df['RFM_Score'] >= 9 and df['R_Score'] >= 4:
        return 'Loyal Customers'
    elif df['RFM_Score'] >= 8 and df['R_Score'] >= 4:
        return 'Potential Loyalists'
    elif df['R_Score'] >= 4 and df['F_Score'] <= 2:
        return 'New Customers'
    elif df['R_Score'] >= 3 and df['RFM_Score'] >= 6:
        return 'Promising'
    elif df['R_Score'] >= 3 and df['M_Score'] >= 4:
        return 'Need Attention'
    elif df['R_Score'] <= 2 and df['RFM_Score'] >= 6:
        return 'At Risk'
    elif df['R_Score'] <= 2 and df['F_Score'] >= 3:
        return 'Cannot Lose Them'
    elif df['R_Score'] <= 2 and df['RFM_Score'] <= 5:
        return 'Hibernating'
    else:
        return 'Lost'

rfm['Customer_Segment'] = rfm.apply(segment_customers, axis=1)

print("\nCustomer segments created!")
print("\n" + "="*60)
print("SEGMENT DISTRIBUTION:")
print("="*60)
segment_counts = rfm['Customer_Segment'].value_counts()
print(segment_counts)

print("\n" + "="*60)
print("SEGMENT STATISTICS:")
print("="*60)
segment_stats = rfm.groupby('Customer_Segment').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean'
}).round(2)
print(segment_stats)

# ============================================
# SAVE RFM DATA
# ============================================
rfm.to_csv('data/rfm_analysis.csv')
print("\n" + "="*60)
print("RFM analysis saved to 'data/rfm_analysis.csv'")
print("="*60)

print("\n✅ RFM ANALYSIS COMPLETE!")
```

```python
# ============================================
# PHASE 4: DATA VISUALIZATION
# ============================================

import matplotlib.pyplot as plt
import seaborn as sns

print("="*60)
print("CREATING VISUALIZATIONS...")
print("="*60)

# Set style for better-looking plots
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# ============================================
# VISUALIZATION 1: Customer Segment Distribution
# ============================================
plt.figure(figsize=(12, 6))
segment_counts = rfm['Customer_Segment'].value_counts()
colors = sns.color_palette("husl", len(segment_counts))

plt.subplot(1, 2, 1)
segment_counts.plot(kind='bar', color=colors)
plt.title('Customer Segment Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Customer Segment', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.subplot(1, 2, 2)
plt.pie(segment_counts, labels=segment_counts.index, autopct='%1.1f%%', 
        colors=colors, startangle=90)
plt.title('Customer Segment Percentage', fontsize=14, fontweight='bold')
plt.tight_layout()

plt.savefig('images/segment_distribution.png', dpi=300, bbox_inches='tight')
print("✅ Saved: segment_distribution.png")
plt.show()

# ============================================
# VISUALIZATION 2: RFM Heatmap by Segment
# ============================================
plt.figure(figsize=(10, 6))
segment_rfm = rfm.groupby('Customer_Segment')[['Recency', 'Frequency', 'Monetary']].mean()

sns.heatmap(segment_rfm.T, annot=True, fmt='.0f', cmap='YlOrRd', 
            cbar_kws={'label': 'Average Value'})
plt.title('Average RFM Values by Customer Segment', fontsize=14, fontweight='bold')
plt.xlabel('Customer Segment', fontsize=12)
plt.ylabel('RFM Metrics', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.savefig('images/rfm_heatmap.png', dpi=300, bbox_inches='tight')
print("✅ Saved: rfm_heatmap.png")
plt.show()

# ============================================
# VISUALIZATION 3: RFM Score Distribution
# ============================================
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Recency Distribution
axes[0].hist(rfm['Recency'], bins=50, color='skyblue', edgecolor='black')
axes[0].set_title('Recency Distribution', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Days Since Last Purchase', fontsize=10)
axes[0].set_ylabel('Number of Customers', fontsize=10)

# Frequency Distribution
axes[1].hist(rfm['Frequency'], bins=50, color='lightcoral', edgecolor='black')
axes[1].set_title('Frequency Distribution', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Number of Purchases', fontsize=10)
axes[1].set_ylabel('Number of Customers', fontsize=10)

# Monetary Distribution
axes[2].hist(rfm['Monetary'], bins=50, color='lightgreen', edgecolor='black')
axes[2].set_title('Monetary Distribution', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Total Spend ($)', fontsize=10)
axes[2].set_ylabel('Number of Customers', fontsize=10)

plt.tight_layout()
plt.savefig('images/rfm_distributions.png', dpi=300, bbox_inches='tight')
print("✅ Saved: rfm_distributions.png")
plt.show()

# ============================================
# VISUALIZATION 4: Top 10 Customers by Monetary Value
# ============================================
plt.figure(figsize=(10, 6))
top_customers = rfm.nlargest(10, 'Monetary')[['Monetary', 'Frequency']]

plt.barh(range(len(top_customers)), top_customers['Monetary'], color='gold', edgecolor='black')
plt.yticks(range(len(top_customers)), [f'Customer {i+1}' for i in range(len(top_customers))])
plt.xlabel('Total Spend ($)', fontsize=12)
plt.ylabel('Customer', fontsize=12)
plt.title('Top 10 Customers by Total Spend', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()

plt.savefig('images/top_customers.png', dpi=300, bbox_inches='tight')
print("✅ Saved: top_customers.png")
plt.show()

# ============================================
# VISUALIZATION 5: Segment Revenue Contribution
# ============================================
plt.figure(figsize=(12, 6))
segment_revenue = rfm.groupby('Customer_Segment')['Monetary'].sum().sort_values(ascending=False)

plt.subplot(1, 2, 1)
segment_revenue.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Total Revenue by Segment', fontsize=14, fontweight='bold')
plt.xlabel('Customer Segment', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.subplot(1, 2, 2)
colors = sns.color_palette("Set2", len(segment_revenue))
plt.pie(segment_revenue, labels=segment_revenue.index, autopct='%1.1f%%', 
        colors=colors, startangle=90)
plt.title('Revenue Contribution by Segment', fontsize=14, fontweight='bold')
plt.tight_layout()

plt.savefig('images/segment_revenue.png', dpi=300, bbox_inches='tight')
print("✅ Saved: segment_revenue.png")
plt.show()

# ============================================
# VISUALIZATION 6: 3D Scatter Plot (RFM)
# ============================================
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Create color map for segments
segment_colors = {segment: color for segment, color in 
                 zip(rfm['Customer_Segment'].unique(), 
                     sns.color_palette("husl", rfm['Customer_Segment'].nunique()))}

for segment in rfm['Customer_Segment'].unique():
    segment_data = rfm[rfm['Customer_Segment'] == segment]
    ax.scatter(segment_data['Recency'], 
              segment_data['Frequency'], 
              segment_data['Monetary'],
              c=[segment_colors[segment]], 
              label=segment, 
              alpha=0.6, 
              s=50)

ax.set_xlabel('Recency (Days)', fontsize=10)
ax.set_ylabel('Frequency (Orders)', fontsize=10)
ax.set_zlabel('Monetary ($)', fontsize=10)
ax.set_title('3D RFM Analysis by Customer Segment', fontsize=14, fontweight='bold')
ax.legend(loc='upper left', fontsize=8)

plt.tight_layout()
plt.savefig('images/rfm_3d_scatter.png', dpi=300, bbox_inches='tight')
print("✅ Saved: rfm_3d_scatter.png")
plt.show()

print("\n" + "="*60)
print("✅ ALL VISUALIZATIONS CREATED SUCCESSFULLY!")
print("="*60)
print("\nImages saved in 'images/' folder:")
print("1. segment_distribution.png")
print("2. rfm_heatmap.png")
print("3. rfm_distributions.png")
print("4. top_customers.png")
print("5. segment_revenue.png")
print("6. rfm_3d_scatter.png")
```
